<h1 style="color:Gray">Inferential Statistics</h1>

```python
# Libraries
import numpy as np
import pandas as pd

# Viz
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 

# Stats
import scipy.stats as st
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy.stats import norm
```

## Hypothesis:
- **Null hypothesis**: average blood pressure from people in Hawaii is 120 mm Hg
- **Alternative Hypothesis**: average blood pressure from people in Hawaii is higher than 120 mm Hg

```python
# Data
mu = 120 # assumed mean (population mean)
n = 100 # number of sample
std = 21.21 
x = 130.1 # sample mean 
alpha = 0.05
```

```python
# Calculate the t-test
t_stat = (x - mu)/(std/np.sqrt(n))
print(f"T-test (test statistic) = {t_stat:.5f}")
```

```python
# Critical Value
crit_value = st.norm.ppf(alpha)
print(f"Critical Value = {crit_value:.2f}")
```

```python
# P_Value
p_value = 1 - norm.cdf(t_stat)
print(f"P_Value = {p_value:.7f}")

# decision making
if p_value < alpha:
    print("Reject Null Hypothesis")
else:
    print("Fail to reject Null Hypothesis")
```

```python
# Viz
x = np.linspace(-3, 3, 100)
y = st.norm.pdf(x)
plt.plot(x, y, label="Standard Normal Distribution")
plt.fill_between(x, 0, y, where=(x < crit_value), color='red', alpha=0.3, label='Rejection Region')
plt.axvline(t_stat, color='blue', linestyle='--', label=f'Test Statistic (t = {t_stat:.2f})')
plt.axvline(crit_value, color='red', linestyle='--', label=f'Critical Value (z = {crit_value:.2f})')
plt.title("Hypothesis Testing Visualization")
plt.legend(loc='lower center')
plt.show()
```
